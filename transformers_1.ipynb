```python
### Task: Image captioning
captioner = pipeline(task="image-to-text",
                     model="nlpconnect/vit-gpt2-image-captioning")
```

```python
file_name = file_name1
```

```python
captioner2 = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")
```

```python
captioner3 = pipeline("image-to-text", model="Salesforce/blip2-opt-2.7b")
```

```python
print(captioner(file_name))
print(captioner2(file_name))
print(captioner3(file_name))
display(Image(filename='cat.jpg'))
```

```python
print(captioner(file_name2))
print(captioner2(file_name2))
print(captioner3(file_name2))
display(Image(file_name2))
```
